```python
import pandas as pd
```

```python
hotel=pd.read_csv('Hotel_Reviews.csv')
hotel.head()
```

```python
ukhotel=hotel.loc[hotel['Reviewer_Nationality']==' United Kingdom ']
len(ukhotel)
```

```python
ukhotel.reset_index(drop=True,inplace=True)
```

```python
from typing import List
def lower(lines: List[str]):
    '''return all characters in lowercase'''
    return [line.lower() for line in lines]
```

```python
ukhotel['Lower_Neg_Review'] = lower(ukhotel['Negative_Review'])
ukhotel['Lower_Pos_Review'] = lower(ukhotel['Positive_Review'])
```

```python
# Find the most common words
from collections import Counter
import nltk
from nltk.corpus import stopwords
counter=Counter()
for review in ukhotel['Lower_Pos_Review']:
    words = nltk.word_tokenize(review)  # "words" still contains punctuations
    for word in words:
        counter[word]+=1
top100pos=counter.most_common(100)
top100pos
```

```python
# Find the most common words
for review in ukhotel['Lower_Neg_Review']:
    words = nltk.word_tokenize(review)  # "words" still contains punctuations
    for word in words:
        counter[word]+=1
top100neg=counter.most_common(100)
top100neg
```

```python
len(ukhotel[ukhotel['days_since_review']<365])
```

```python
uk=ukhotel[ukhotel['days_since_review']<365]
uk.reset_index(drop=True,inplace=True)
uk.head()
```

```python
from textacy.preprocessing.replace import urls, hashtags, numbers, emails, emojis, currency_symbols
uk["Positive_Review_Clean"] = uk['Lower_Pos_Review'].\
  apply(urls).\
  apply(hashtags).\
  apply(numbers).\
  apply(currency_symbols).\
  apply(emojis).\
  apply(emails)
uk["Negative_Review_Clean"] = uk['Lower_Neg_Review'].\
  apply(urls).\
  apply(hashtags).\
  apply(numbers).\
  apply(currency_symbols).\
  apply(emojis).\
  apply(emails)
uk.head(5)
```

```python
from textacy.preprocessing.remove import punctuation
uk['Positive_Review_Clean'].apply(punctuation)
uk['Negative_Review_Clean'].apply(punctuation)
uk.head()
```

```python
uk.to_csv('uk_hotel.csv',index=False)
```
